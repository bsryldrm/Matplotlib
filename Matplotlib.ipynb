```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
yasListesi = [10,20,30,40,50,60,70]
kiloListesi = [40,60,80,85,86,87,70]
```

```python
npYasListesi = np.array(yasListesi)
npKiloListesi = np.array(kiloListesi)
```

```python
plt.plot(npYasListesi,npKiloListesi,"b")    # b ---> blue renk 
plt.xlabel("Yas")    # x eksenine yaş başlığı at
plt.ylabel("Kilo")   # y eksenine kilo başlığı at
plt.title("Kilo'nun yaşa göre değişim grafiği")  # ana başlık
```

```python
numpyDizisi = np.linspace(0,10,20)
numpyDizisi
```

```python
numpyDizisi2 = numpyDizisi**3
numpyDizisi2
```

```python
plt.plot(numpyDizisi,numpyDizisi2,"g--")        # çizgilerin kesikli gösterimi.. 
```

```python
plt.plot(numpyDizisi,numpyDizisi2,"g*-")         # bir farklı gösterim..
```

```python
plt.subplot(1,2,1)
plt.plot(numpyDizisi,numpyDizisi2,"r*-")
plt.subplot(1,2,2)  
plt.plot(numpyDizisi,numpyDizisi2,"g--")
```

```python
# Özelleştirme
figure = plt.figure()
figureAxes = figure.add_axes([0.4,0.4,0.5,0.5])
figureAxes.plot(numpyDizisi,numpyDizisi2,"g")
figureAxes.set_xlabel("X ekseni veri ismi")
figureAxes.set_ylabel("Y ekseni veri ismi")
figureAxes.set_title("Grafik Başlığı")
```

```python
# iç içe grafikler
figure2 = plt.figure()

eksen1 = figure2.add_axes([0.1,0.1,0.7,0.7])
eksen2 = figure2.add_axes([0.2,0.3,0.3,0.3])

eksen1.plot(numpyDizisi,numpyDizisi2,"g")
eksen1.set_xlabel("X ekseni")
eksen1.set_ylabel("Y ekseni")
eksen1.set_title("Ana Grafik Başlık")

eksen2.plot(numpyDizisi2,numpyDizisi,"b")
eksen2.set_xlabel("X ekseni küçük")
eksen2.set_ylabel("Y ekseni küçük")
eksen2.set_title("Küçük Grafik Başlık")
```

```python
plt.subplots()
```

```python
(Figure1, eksen1) = plt.subplots()
eksen1.plot(numpyDizisi,numpyDizisi2,"g--")
```

```python
(Figure1, eksen1) = plt.subplots(nrows = 1, ncols=2)
type(eksen1)
```

```python
(Figure1, eksen1) = plt.subplots(nrows = 1, ncols=2)
type(eksen1)

for eksen in eksen1:
    eksen.plot(numpyDizisi,numpyDizisi2,"g")
    eksen.set_xlabel("X ekseni")
    
plt.tight_layout()   # grafikler arasında ki mesafeyi artırır
```

```python
yeniFigur = plt.figure()

yeniEksen = yeniFigur.add_axes([0.1,0.1,0.9,0.9])

yeniEksen.plot(numpyDizisi, numpyDizisi**2, label = "Numpy Dizisi ** 2")
yeniEksen.plot(numpyDizisi, numpyDizisi**3, label = "Numpy Dizisi ** 3")
yeniEksen.legend(loc = 2)    # loc=2  bilgi ekranını grafiğin sol üst köşesine yazdırır. sayı değiştirerek farklı köşelere yazdırılabilir.
```
